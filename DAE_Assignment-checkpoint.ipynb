```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("C:/Users/Duma/Downloads/lawyers.csv",skiprows=0)
data.head(141)
```

```python
data.plot.scatter(x='CLM', y='CTM',color='hotpink')
data.plot.scatter(x='AGE', y='CTM',color='orange')
data.plot.scatter(x='LVL', y='CTM',color='red')
data.plot.scatter(x='SDY', y='CTM')
```

```python
from sklearn.linear_model import LinearRegression

#initiate linear regression model
model = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()

#define predictor and response variables
X = data[["SDY"]]                           #Predictors
y = data["CTM"]                             #target

#define predictor and response variables
A = data[["SDY","CLM"]]                           #Predictors
B = data["CTM"]                             #target

#define predictor and response variables
C = data[["SDY","CLM","AGE"]]                           #Predictors
D = data["CTM"]                             #target

#define predictor and response variables
# E = data[["SDY","CLM","AGE","LVL"]]                           #Predictors
# F = data["CTM"]                             #target

#fit regression model
model.fit(X, y)
model2.fit(A, B)
model3.fit(C, D)

#calculate R-squared of regression model
r_squared = model.score(X, y)
r_squared2 = model2.score(A,B)
r_squared3 = model3.score(C,D)

#view R-squared value
print(r_squared)
print(r_squared2)
print(r_squared3)
```
